```python
import pandas as pd
import numpy as np
import operator
```

```python
pred = pd.read_csv('Results for 10000 trained.csv')
pred.head()
```

```python
col_names = {'AL Dhameen':'AL_Dhameen', 
             'MINHA Costing Analysis & Expense Reduction':'MINHA_Costing_Analysis_and_Expense_Reduction', 
             'Direct Lending':'Direct_Lending', 'Marketing & PR':'Marketing_and_PR', 
             'Business Consultancy and Mentoring':'Business_Consultancy_and_Mentoring', 
             'OQOOD legal services':'OQOOD_legal_services', 'Export Finance':'Export_Finance', 
             'ISTITHMAR (SME EQUITY)':'ISTITHMAR_SME_EQUITY'}

pred.rename(col_names, axis=1, inplace=True)
```

```python
df = pd.read_csv('synthetic_data_latest.csv')
df.head()
```

```python
df.loc[df['Facility product Name'] == "AL Dhameen", 'Facility product Name'] = 'AL_Dhameen'
df.loc[df['Facility product Name'] == "MINHA Costing Analysis & Expense Reduction", 'Facility product Name'] = 'MINHA_Costing_Analysis_and_Expense_Reduction'
df.loc[df['Facility product Name'] == "Direct Lending", 'Facility product Name'] = 'Direct_Lending'
df.loc[df['Facility product Name'] == "Marketing & PR", 'Facility product Name'] = 'Marketing_and_PR'
df.loc[df['Facility product Name'] == "Business Consultancy and Mentoring", 'Facility product Name'] = 'Business_Consultancy_and_Mentoring'
df.loc[df['Facility product Name'] == "OQOOD legal services", 'Facility product Name'] = 'OQOOD_legal_services'
df.loc[df['Facility product Name'] == "Export Finance", 'Facility product Name'] = 'Export_Finance'
df.loc[df['Facility product Name'] == "ISTITHMAR (SME EQUITY)", 'Facility product Name'] = 'ISTITHMAR_SME_EQUITY'
```

```python
def refine_recommendation(rating, op_status, dpd_status, top_n_recommendation_list):
    if ((rating in [1 ,2]) & (op_status in ['Operational', 'Setup']) & (dpd_status == 'Nil') & ('AL_Dhameen' in top_n_recommendation_list)):
        pass
    else:
        if 'AL_Dhameen' in top_n_recommendation_list:
            top_n_recommendation_list.remove('AL_Dhameen')
    if (rating in [1 ,2]) & (op_status in ['Operational', 'Setup']) & (dpd_status == 'Nil') & ('ISTITHMAR (SME_EQUITY)' in top_n_recommendation_list):
        pass
    else:
        if 'ISTITHMAR (SME_EQUITY)' in top_n_recommendation_list:
            top_n_recommendation_list.remove('ISTITHMAR (SME_EQUITY)')
    if (rating in [1 ,2 ,3]) & (op_status in ['Operational', 'Setup']) & (dpd_status == 'Nil') & ('Export_Finance' in top_n_recommendation_list):
        pass
    else: 
        if 'Export_Finance' in top_n_recommendation_list:
            top_n_recommendation_list.remove('Export_Finance')
    if ((rating in [1 ,2]) & (op_status in ['Operational', 'Setup']) & (dpd_status == 'Nil') & ('Direct_Lending' in top_n_recommendation_list)):
        pass
    else: 
        if 'Direct_Lending' in top_n_recommendation_list:
            top_n_recommendation_list.remove('Direct_Lending')
    if ((rating in [1 ,2 ,3 ,4 ,5]) & (op_status in ['Operational', 'Setup']) & (dpd_status in ['Nil', 'Closed', 'Defaulted']) & ('Business_Consultancy_and_Mentoring' in top_n_recommendation_list)):
        pass
    else: 
        if 'Business_Consultancy_and_Mentoring' in top_n_recommendation_list:
            top_n_recommendation_list.remove('Business_Consultancy_and_Mentoring')
    if ((rating in [1 ,2 ,3 ,4 ,5]) & (op_status in ['Operational', 'Setup', 'Temporarily Closed', 'Cancelled']) & (dpd_status in ['Nil', 'Closed', 'Defaulted']) & ('Marketing_and_PR' in top_n_recommendation_list)):
        pass
    else:
        if 'Marketing_and_PR' in top_n_recommendation_list:
            top_n_recommendation_list.remove('Marketing_and_PR')
    if ((rating in [1 ,2 ,3 ,4 ,5]) & (op_status in ['Operational', 'Setup', 'Temporarily Closed', 'Permanently Closed']) & (dpd_status in ['Nil', 'Closed', 'Defaulted']) & ('MINHA_Costing_Analysis_and_Expense_Reduction' in top_n_recommendation_list)):
        pass
    else:
        if 'MINHA_Costing_Analysis_and_Expense_Reduction' in top_n_recommendation_list:
            top_n_recommendation_list.remove('MINHA_Costing_Analysis_and_Expense_Reduction')
    
    return top_n_recommendation_list
```

```python
def gbdt_product_recommender(cif_num, top_n):
    pred_sub = pred[pred['CIF_Number']==cif_num]
    for index, rows in pred_sub.iterrows():
        my_list = {'EYADA':rows.EYADA, 'AL_Dhameen':rows.AL_Dhameen, 
                   'MINHA_Costing_Analysis_and_Expense_Reduction':rows.MINHA_Costing_Analysis_and_Expense_Reduction, 
                   'Direct_Lending':rows.Direct_Lending, 'Marketing_and_PR':rows.Marketing_and_PR, 
                   'Business_Consultancy_and_Mentoring':rows.Business_Consultancy_and_Mentoring, 
                   'OQOOD_legal_services':rows.OQOOD_legal_services, 'Export_Finance':rows.Export_Finance, 
                   'ISTITHMAR (SME_EQUITY)':rows.ISTITHMAR_SME_EQUITY}
        sorted_my_list = dict( sorted(my_list.items(), key=operator.itemgetter(1), reverse=True))
        top_n_recommendation = list(sorted_my_list.keys())[0:top_n]
        # print(f"Products recommended before refining: {top_n_recommendation}")
        df_sub = df[df['CIF Number']==cif_num]
        index_list = df[df['CIF Number']==cif_num].index.tolist()
        new_recommendations = []
        for i in index_list:
            rating = df_sub.at[i, 'Match as per rating']
            op_status = df_sub.at[i, 'Match as per Operational status']
            dpd_status = df_sub.at[i, 'Match as per DPD Status']
            new_top_n_recommendation = refine_recommendation(rating, op_status, dpd_status, top_n_recommendation_list=top_n_recommendation)
            top_n_recommendation = list(sorted_my_list.keys())[0:top_n]
            for j in new_top_n_recommendation:
                new_recommendations.append(j)
        new_recommendations = list(set(new_recommendations))
        # print(f"Products recommended after refining: {new_recommendations}")
    return new_recommendations
```

```python
inference_df = pd.DataFrame(columns = ['Company_Name', 'CIF_Number', 'Ground_Truth', 'Recommended_Matches', 'Count'])
inference_df['Company_Name'] = pred['Company_Name']
inference_df['CIF_Number'] = pred['CIF_Number']
```

```python
product_gt = []
for num in list(pred['CIF_Number']):
    df1 = df[df['CIF Number']==num]
    sub_list = list(df1['Facility product Name'])
    sub_list = list(set(sub_list))
    product_gt.append(sub_list)

inference_df['Ground_Truth'] = product_gt
```

```python
for i in range(len(pred)):
    cif_num = inference_df.at[i, 'CIF_Number']
    inference_df.at[i, 'Recommended_Matches'] = gbdt_product_recommender(cif_num, top_n=5)
```

```python
inference_df
```

```python
for i, index in zip(inference_df['Ground_Truth'], range(len(inference_df))):
    count = []
    for j in i:
        if j in inference_df.at[index, 'Recommended_Matches']:
            count.append(1)
        else:
            count.append(0)
    inference_df.at[index, 'Count'] = count
```

```python
count = 0
for i in range(len(inference_df)):
    if 1 in inference_df.at[i, 'Count']:
        count += 1
count
```

```python
accuracy = count/len(inference_df)
print(f"Accuracy: {accuracy*100}%")
```

```python
inference_df.to_csv('Top5 Recommendations.csv', index=False)
```
